## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
Create a Python program that:
- Asks the user for their monthly income.
- Allows the user to input up to 3 individual expenses. For each expense, it should ask for the expense name and the expense amount.
- Calculates the remaining budget by subtracting the total expenses from the monthly income.
- Prints the total expenses.
- Prints the remaining budget.
- Displays a clear warning message if the remaining budget is negative (e.g., 'Warning: You are over budget!').
- Ensure all monetary inputs are handled as floating-point numbers and formatted to two decimal places in the output.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE

# Get monthly income
while True:
    try:
        monthly_income = float(input("Enter your monthly income: "))
        if monthly_income < 0:
            print("Income cannot be negative. Please enter a positive value.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value for income.")

expenses = []
num_expenses = 0
MAX_EXPENSES = 3

# Get up to 3 expenses
while num_expenses < MAX_EXPENSES:
    add_more = input(f"Do you want to add an expense? (yes/no) You can add up to {MAX_EXPENSES - num_expenses} more: ").lower()
    if add_more != 'yes':
        break

    expense_name = input(f"Enter name for expense {num_expenses + 1}: ")
    expense_category = input(f"Enter category for {expense_name} (e.g., Food, Transport, Rent): ")
    while True:
        try:
            expense_amount = float(input(f"Enter amount for {expense_name}: "))
            if expense_amount < 0:
                print("Expense amount cannot be negative. Please enter a positive value.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a numerical value for the expense amount.")

    expenses.append({'name': expense_name, 'amount': expense_amount, 'category': expense_category})
    num_expenses += 1

# Calculate total expenses
total_expenses = sum(expense['amount'] for expense in expenses)

# Calculate remaining budget
remaining_budget = monthly_income - total_expenses

# Display results
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:.2f}")
print("Expenses:")
if expenses:
    for expense in expenses:
        print(f"  - {expense['name']} (Category: {expense['category']}): ${expense['amount']:.2f}")
else:
    print("  No expenses added.")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Show warning if budget is negative
if remaining_budget < 0:
    print("\nWarning: You are over budget! Consider reducing your expenses.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: categories"
enhancement_prompt = """
I asked for the code to be more specific and detailed on the information required for expenses, which led to the addition of expense categories for each entry.
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER

import os

# --- Constants ---
DATA_FILENAME = 'budget_data.txt'
MAX_EXPENSES_PER_SESSION = 3 # Still limit adding expenses per session for simplicity

# --- Helper Functions for Input and Validation ---
def get_validated_float_input(prompt, error_message="Invalid input. Please enter a numerical value.", allow_negative=False):
    while True:
        try:
            value = float(input(prompt))
            if not allow_negative and value < 0:
                print("Value cannot be negative. Please enter a positive value.")
                continue
            return value
        except ValueError:
            print(error_message)

def get_string_input(prompt, error_message="Input cannot be empty. Please try again."):
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print(error_message)

# --- Data Persistence Functions ---
def load_budget_data():
    monthly_income = 0.0
    expenses = []
    if os.path.exists(DATA_FILENAME):
        with open(DATA_FILENAME, 'r') as f:
            lines = f.readlines()
            if lines:
                try:
                    monthly_income = float(lines[0].strip())
                    for line in lines[1:]:
                        parts = line.strip().split(',')
                        if len(parts) == 3:
                            expenses.append({
                                'name': parts[0],
                                'amount': float(parts[1]),
                                'category': parts[2]
                            })
                except (ValueError, IndexError) as e:
                    print(f"Error loading data: {e}. Starting with empty budget.")
                    monthly_income = 0.0
                    expenses = []
    return monthly_income, expenses

def save_budget_data(monthly_income, expenses):
    with open(DATA_FILENAME, 'w') as f:
        f.write(f"{monthly_income:.2f}\n")
        for expense in expenses:
            f.write(f"{expense['name']},{expense['amount']:.2f},{expense['category']}\n")
    print(f"Budget data saved to {DATA_FILENAME}")

# --- Budget Management Functions ---
def set_monthly_income():
    return get_validated_float_input("Enter your monthly income: ")

def add_expense(expenses_list):
    expense_name = get_string_input(f"Enter name for expense (e.g., Groceries, Rent): ")
    expense_category = get_string_input(f"Enter category for {expense_name} (e.g., Food, Transport, Rent): ")
    expense_amount = get_validated_float_input(f"Enter amount for {expense_name}: ")
    expenses_list.append({'name': expense_name, 'amount': expense_amount, 'category': expense_category})
    print(f"Expense '{expense_name}' added successfully.")

def calculate_budget_metrics(monthly_income, expenses_list):
    total_expenses = sum(expense['amount'] for expense in expenses_list)
    remaining_budget = monthly_income - total_expenses
    return total_expenses, remaining_budget

def display_detailed_summary(monthly_income, expenses_list):
    print("\n--- Budget Summary ---")
    print(f"Monthly Income: ${monthly_income:.2f}")

    if not expenses_list:
        print("No expenses added yet.")
        total_expenses = 0.0
        remaining_budget = monthly_income
    else:
        print("Expenses:")
        for expense in expenses_list:
            print(f"  - {expense['name']} (Category: {expense['category']}): ${expense['amount']:.2f}")

        total_expenses, remaining_budget = calculate_budget_metrics(monthly_income, expenses_list)
        print(f"\nTotal Expenses: ${total_expenses:.2f}")
        print(f"Remaining Budget: ${remaining_budget:.2f}")

        # Category-wise spending
        print("\n--- Spending by Category ---")
        category_spending = {}
        for expense in expenses_list:
            category = expense['category'].strip().lower() # Normalize category names
            category_spending[category] = category_spending.get(category, 0.0) + expense['amount']

        for category, amount in category_spending.items():
            print(f"  - {category.capitalize()}: ${amount:.2f}")

        # Show warning if budget is negative
        if remaining_budget < 0:
            print("\nWarning: You are over budget! Consider reducing your expenses.")
        else:
            print("\nGood job! You are within your budget.")

    # Percentage of income spent
    if monthly_income > 0:
        percentage_spent = (total_expenses / monthly_income) * 100
        print(f"You have spent {percentage_spent:.2f}% of your monthly income.")

# --- Main Program Loop ---
def main():
    monthly_income, expenses = load_budget_data()
    print("\nWelcome to your Enhanced Budget Tracker!")

    while True:
        print("\n--- Main Menu ---")
        print("1. Set/Update Monthly Income")
        print("2. Add Expense")
        print("3. View Budget Summary")
        print("4. Save Budget Data")
        print("5. Exit")

        choice = get_string_input("Enter your choice (1-5): ")

        if choice == '1':
            monthly_income = set_monthly_income()
        elif choice == '2':
            # Allow adding multiple expenses in one go, up to MAX_EXPENSES_PER_SESSION
            expenses_added_this_session = 0
            while expenses_added_this_session < MAX_EXPENSES_PER_SESSION:
                add_more = get_string_input("Do you want to add an expense? (yes/no): ").lower()
                if add_more == 'yes':
                    add_expense(expenses)
                    expenses_added_this_session += 1
                else:
                    break
        elif choice == '3':
            display_detailed_summary(monthly_income, expenses)
        elif choice == '4':
            save_budget_data(monthly_income, expenses)
        elif choice == '5':
            save_budget_data(monthly_income, expenses) # Save before exiting
            print("Exiting Budget Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == '__main__':
    main()
```

## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    # Your validation and storage code here
    try:
        expense_amount = float(expense_input)
        if expense_amount <= 0:
            print("Expense amount must be positive.")
            continue

        expense_description = input("What was this expense for? ")
        if not expense_description:
            print("Description cannot be empty. Please try again.")
            continue

        expenses.append(expense_amount)
        descriptions.append(expense_description)
        total_spent += expense_amount

        budget_percentage_used = (total_spent / monthly_budget) * 100
        print(f"Current total spent: ${total_spent:.2f} ({budget_percentage_used:.1f}% of budget)")

    except ValueError:
        print("Invalid input. Please enter a valid number for the expense.")


print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Your task: Find highest and lowest expenses
    # Use the min/max pattern from workshops
    highest_expense = max(expenses)
    lowest_expense = min(expenses)

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Your task: Categorize expenses by size
    # Count how many expenses fall into each category:
    # Large (>20% of budget), Medium (5-20%), Small (<5%)

    large_count = 0    # > 20% of budget
    medium_count = 0   # 5-20% of budget
    small_count = 0    # < 5% of budget

    budget_20_percent = monthly_budget * 0.20
    budget_5_percent = monthly_budget * 0.05

    for expense in expenses:
        if expense > budget_20_percent:
            large_count += 1
        elif expense >= budget_5_percent: # Between 5% and 20% (inclusive of 5%)
            medium_count += 1
        else: # Less than 5%
            small_count += 1

    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # Your task: Provide smart feedback based on the analysis
    # Give different advice based on spending patterns
    print("\n💡 Smart Feedback:")

    if total_spent > monthly_budget:
        print("🚨 Warning! You have exceeded your monthly budget!")
        print(f"Consider reviewing your spending, you are over by ${total_spent - monthly_budget:.2f}.")
    elif (total_spent / monthly_budget) * 100 > 80:
        print("⚠️ You are close to your budget limit. Keep an eye on your remaining spending!")
    elif large_count > 0:
        print(f"Consider if your {large_count} large expense(s) were truly necessary this month.")
    else:
        print("👍 Great job managing your expenses! Your spending is well within budget.")

    if monthly_budget - total_spent < 0:
        print(f"You are ${abs(monthly_budget - total_spent):.2f} over budget.")
    elif monthly_budget - total_spent > 0:
        print(f"You have ${monthly_budget - total_spent:.2f} remaining in your budget.")
    else:
        print("You have spent exactly your budget this month.")


else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Get goal name
    goal_name = input("Enter goal name: ")
    while not goal_name:
        print("Goal name cannot be empty. Please try again.")
        goal_name = input("Enter goal name: ")

    # Get target amount with validation
    target_amount = 0.0
    while True:
        try:
            target_amount = float(input(f"Enter target amount for '{goal_name}': $"))
            if target_amount <= 0:
                print("Target amount must be positive.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    # Get current amount with validation
    current_amount = -1.0 # Initialize with an invalid value
    while True:
        try:
            current_amount = float(input(f"Enter current amount saved for '{goal_name}': $"))
            if current_amount < 0:
                print("Current amount cannot be negative.")
            elif current_amount > target_amount:
                print("Current amount cannot exceed target amount.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    goal_names.append(goal_name)
    goal_targets.append(target_amount)
    goal_current.append(current_amount)

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0
    min_progress = 101 # Initialize with a value higher than 100%
    goal_needing_attention_index = -1

    for i in range(len(goal_names)):
        # Your progress calculation here
        goal_name = goal_names[i]
        target = goal_targets[i]
        current = goal_current[i]

        progress_percentage = (current / target) * 100 if target > 0 else 0
        amount_remaining = target - current

        months_to_completion = 0
        if amount_remaining > 0:
            months_to_completion = amount_remaining / 100 # Assuming saving $100/month

        print(f"\nGoal: {goal_name}")
        print(f"  Target: ${target:.2f}, Current: ${current:.2f}")
        print(f"  Progress: {progress_percentage:.1f}% complete")
        print(f"  Remaining: ${amount_remaining:.2f}")
        if months_to_completion > 0:
            print(f"  Est. months to completion (saving $100/month): {months_to_completion:.1f}")
        elif amount_remaining <= 0:
            print("  Goal achieved! 🎉")

        total_target += target
        total_current += current

        if progress_percentage < min_progress and amount_remaining > 0:
            min_progress = progress_percentage
            goal_needing_attention_index = i

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Your task: Find which goal needs most attention
    # Find the goal with lowest progress percentage
    # Provide encouragement and specific advice
    print("\n💡 Goals Feedback:")
    if goal_needing_attention_index != -1:
        goal_name_attention = goal_names[goal_needing_attention_index]
        print(f"Your goal '{goal_name_attention}' has the lowest progress ({min_progress:.1f}%).")
        print("Keep going! Small regular contributions can make a big difference over time.")
    else:
        print("All your goals are looking great, or already achieved! Fantastic work!")


else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Display expense summary
        print("\n--- Expense Summary ---")
        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i in range(len(expenses)):
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"\nTotal expenses: ${sum(expenses):.2f}")

    elif choice == "2":
        # Display goals progress
        print("\n--- Goals Progress ---")
        if len(goal_names) == 0:
            print("No savings goals set yet.")
        else:
            for i in range(len(goal_names)):
                goal_name = goal_names[i]
                target = goal_targets[i]
                current = goal_current[i]
                progress_percentage = (current / target) * 100 if target > 0 else 0
                print(f"{goal_name}: ${current:.2f}/${target:.2f} ({progress_percentage:.1f}% complete)")

    elif choice == "3":
        # Add single new expense
        print("\n--- Add New Expense ---")

        while True:
            expense_input = input("Enter new expense amount: $")
            try:
                expense_amount = float(expense_input)
                if expense_amount <= 0:
                    print("Expense amount must be positive.")
                    continue
                break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        expense_description = input("What was this expense for? ")
        while not expense_description:
            print("Description cannot be empty. Please try again.")
            expense_description = input("What was this expense for? ")

        expenses.append(expense_amount)
        descriptions.append(expense_description)
        total_spent += expense_amount # Update global total_spent

        budget_percentage_used = (total_spent / monthly_budget) * 100
        print(f"Expense added! Current total spent: ${total_spent:.2f} ({budget_percentage_used:.1f}% of budget)")

    elif choice == "4":
        # Budget status check
        print("\n--- Budget Check ---")
        budget_percentage_used = (total_spent / monthly_budget) * 100
        remaining_budget = monthly_budget - total_spent

        print(f"Monthly Budget: ${monthly_budget:.2f}")
        print(f"Total Spent: ${total_spent:.2f}")
        print(f"Budget Used: {budget_percentage_used:.1f}%")
        print(f"Remaining Budget: ${remaining_budget:.2f}")

        if remaining_budget < 0:
            print("🚨 Warning! You are over budget! Consider cutting back on non-essential spending.")
        elif budget_percentage_used > 80:
            print("⚠️ You are close to your budget limit. Spend wisely for the rest of the month.")
        else:
            print("👍 Your spending is well within budget. Keep up the good work!")

    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```

```python
import sys
from io import StringIO

# Define a class to mock input
class MockInput:
    def __init__(self, inputs):
        self.inputs = inputs
        self.current_input = 0

    def readline(self):
        if self.current_input < len(self.inputs):
            response = self.inputs[self.current_input]
            self.current_input += 1
            # print(f"Simulating input: {response}") # Uncomment to see simulated inputs
            return response + '\n'
        else:
            # If more inputs are requested than provided, return empty string or raise EOFError
            return ''

# Prepare the mock inputs for the test scenario
mock_inputs = [
    'Test User',      # Name
    '2000',           # Budget
    '2',              # How many savings goals?
    'Emergency Fund', # Goal 1 name
    '1000',           # Goal 1 target
    '250',            # Goal 1 current
    'Vacation',       # Goal 2 name
    '500',            # Goal 2 target
    '50',             # Goal 2 current
    '450',            # Expense 1 amount
    'Rent payment',   # Expense 1 description
    '67.50',          # Expense 2 amount
    'Weekly groceries', # Expense 2 description
    '89.00',          # Expense 3 amount
    'Gas for car',    # Expense 3 description
    'done',           # Finish expense entry
    '1',              # Menu choice: View expense summary
    '2',              # Menu choice: View goals progress
    '3',              # Menu choice: Add new expense
    '25',             # New expense amount
    'Coffee',         # New expense description
    '4',              # Menu choice: Budget check
    '5'               # Menu choice: Exit
]

# Create a StringIO object to capture print output
captured_output = StringIO()

# Store original stdin and stdout
original_stdin = sys.stdin
original_stdout = sys.stdout

# Redirect stdin and stdout
sys.stdin = MockInput(mock_inputs)
sys.stdout = captured_output

try:
    # --- Start of Combined Finance Tracker Code ---

    print("=== Enhanced Finance Tracker v0.3 ===")
    print("Week 3: Loop-Powered Edition")
    print()

    # User setup (enhanced from Week 2)
    name = input("Enter your name: ")

    # Add budget validation using while loop
    monthly_budget = 0
    while monthly_budget < 100 or monthly_budget > 10000:
        try:
            budget_input = input("Enter monthly budget ($100-$10,000): $")
            monthly_budget = float(budget_input)

            if monthly_budget < 100:
                print("Budget too low! Minimum $100")
            elif monthly_budget > 10000:
                print("Budget seems high! Maximum $10,000")
        except ValueError:
            print("Please enter a valid number!")

    print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

    # ========== MULTI-EXPENSE ENTRY ==========
    expenses = []  # Store all expense amounts
    descriptions = []  # Store what each expense was for
    total_spent = 0

    print("\n--- Enter Your Expenses ---")
    print("Type 'done' when finished entering expenses")

    while True:
        expense_input = input("\nExpense amount (or 'done'): $")

        if expense_input.lower() == 'done':
            break

        try:
            expense_amount = float(expense_input)
            if expense_amount <= 0:
                print("Expense amount must be positive.")
                continue

            expense_description = input("What was this expense for? ")
            if not expense_description:
                print("Description cannot be empty. Please try again.")
                continue

            expenses.append(expense_amount)
            descriptions.append(expense_description)
            total_spent += expense_amount

            budget_percentage_used = (total_spent / monthly_budget) * 100
            print(f"Current total spent: ${total_spent:.2f} ({budget_percentage_used:.1f}% of budget)")

        except ValueError:
            print("Invalid input. Please enter a valid number for the expense.")


    print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")

    # Expense Analysis Section
    if len(expenses) > 0:
        print("\n📊 Expense Analysis:")

        # Basic statistics using loop patterns from workshops
        print(f"Total expenses: {len(expenses)}")
        print(f"Total spent: ${total_spent:.2f}")
        print(f"Average expense: ${total_spent/len(expenses):.2f}")
        print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
        print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

        # Your task: Find highest and lowest expenses
        # Use the min/max pattern from workshops
        highest_expense = max(expenses)
        lowest_expense = min(expenses)

        print(f"\n💰 Expense Range:")
        print(f"Highest expense: ${highest_expense:.2f}")
        print(f"Lowest expense: ${lowest_expense:.2f}")

        # Your task: Categorize expenses by size
        # Count how many expenses fall into each category:
        # Large (>20% of budget), Medium (5-20%), Small (<5%)

        large_count = 0    # > 20% of budget
        medium_count = 0   # 5-20% of budget
        small_count = 0    # < 5% of budget

        budget_20_percent = monthly_budget * 0.20
        budget_5_percent = monthly_budget * 0.05

        for expense in expenses:
            if expense > budget_20_percent:
                large_count += 1
            elif expense >= budget_5_percent: # Between 5% and 20% (inclusive of 5%)
                medium_count += 1
            else: # Less than 5%
                small_count += 1

        print(f"\n📊 Expense Categories:")
        print(f"Large expenses (>20% budget): {large_count}")
        print(f"Medium expenses (5-20% budget): {medium_count}")
        print(f"Small expenses (<5% budget): {small_count}")

        # Your task: Provide smart feedback based on the analysis
        # Give different advice based on spending patterns
        print("\n💡 Smart Feedback:")

        if total_spent > monthly_budget:
            print("🚨 Warning! You have exceeded your monthly budget!")
            print(f"Consider reviewing your spending, you are over by ${total_spent - monthly_budget:.2f}.")
        elif (total_spent / monthly_budget) * 100 > 80:
            print("⚠️ You are close to your budget limit. Keep an eye on your remaining spending!")
        elif large_count > 0:
            print(f"Consider if your {large_count} large expense(s) were truly necessary this month.")
        else:
            print("👍 Great job managing your expenses! Your spending is well within budget.")

        if monthly_budget - total_spent < 0:
            print(f"You are ${abs(monthly_budget - total_spent):.2f} over budget.")
        elif monthly_budget - total_spent > 0:
            print(f"You have ${monthly_budget - total_spent:.2f} remaining in your budget.")
        else:
            print("You have spent exactly your budget this month.")


    else:
        print("\n📝 No expenses entered - try adding some transactions!")

    print("\n=== Savings Goals Tracker ===")

    # Goal collection
    goal_names = []
    goal_targets = []
    goal_current = []

    # Ask how many goals they want to track
    num_goals = 0
    while num_goals < 1 or num_goals > 5:  # Reasonable limits
        try:
            num_goals = int(input("How many savings goals? (1-5): "))
            if num_goals < 1:
                print("Must have at least 1 goal!")
            elif num_goals > 5:
                print("Maximum 5 goals for this tracker!")
        except ValueError:
            print("Please enter a valid number!")

    print(f"\nEntering {num_goals} savings goals:")

    for i in range(num_goals):
        print(f"\n--- Goal {i+1} ---")

        # Get goal name
        goal_name = input("Enter goal name: ")
        while not goal_name:
            print("Goal name cannot be empty. Please try again.")
            goal_name = input("Enter goal name: ")

        # Get target amount with validation
        target_amount = 0.0
        while True:
            try:
                target_amount = float(input(f"Enter target amount for '{goal_name}': $"))
                if target_amount <= 0:
                    print("Target amount must be positive.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        # Get current amount with validation
        current_amount = -1.0 # Initialize with an invalid value
        while True:
            try:
                current_amount = float(input(f"Enter current amount saved for '{goal_name}': $"))
                if current_amount < 0:
                    print("Current amount cannot be negative.")
                elif current_amount > target_amount:
                    print("Current amount cannot exceed target amount.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        goal_names.append(goal_name)
        goal_targets.append(target_amount)
        goal_current.append(current_amount)

    print(f"\n✅ Saved {len(goal_names)} goals!")

    # Goal Analysis and Progress Tracking
    if len(goal_names) > 0:
        print("\n📊 Goals Progress Analysis:")

        total_target = 0
        total_current = 0
        min_progress = 101 # Initialize with a value higher than 100%
        goal_needing_attention_index = -1

        for i in range(len(goal_names)):
            goal_name = goal_names[i]
            target = goal_targets[i]
            current = goal_current[i]

            progress_percentage = (current / target) * 100 if target > 0 else 0
            amount_remaining = target - current

            months_to_completion = 0
            if amount_remaining > 0:
                months_to_completion = amount_remaining / 100 # Assuming saving $100/month

            print(f"\nGoal: {goal_name}")
            print(f"  Target: ${target:.2f}, Current: ${current:.2f}")
            print(f"  Progress: {progress_percentage:.1f}% complete")
            print(f"  Remaining: ${amount_remaining:.2f}")
            if months_to_completion > 0:
                print(f"  Est. months to completion (saving $100/month): {months_to_completion:.1f}")
            elif amount_remaining <= 0:
                print("  Goal achieved! 🎉")

            total_target += target
            total_current += current

            if progress_percentage < min_progress and amount_remaining > 0:
                min_progress = progress_percentage
                goal_needing_attention_index = i

        # Overall progress summary
        overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
        total_remaining = total_target - total_current

        print(f"\n🎯 Overall Goals Summary:")
        print(f"Total target: ${total_target:.2f}")
        print(f"Total saved: ${total_current:.2f}")
        print(f"Overall progress: {overall_progress:.1f}%")
        print(f"Total remaining: ${total_remaining:.2f}")

        # Your task: Find which goal needs most attention
        # Find the goal with lowest progress percentage
        # Provide encouragement and specific advice
        print("\n💡 Goals Feedback:")
        if goal_needing_attention_index != -1:
            goal_name_attention = goal_names[goal_needing_attention_index]
            print(f"Your goal '{goal_name_attention}' has the lowest progress ({min_progress:.1f}%).")
            print("Keep going! Small regular contributions can make a big difference over time.")
        else:
            print("All your goals are looking great, or already achieved! Fantastic work!")


    else:
        print("No goals to analyze!")

    print("\n=== Interactive Finance Menu ===")

    # Simple menu system
    running = True

    while running:
        print("\n" + "-"*30)
        print(f"💰 {name}'s Finance Tracker")
        if len(expenses) > 0:
            print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
        print("-"*30)
        print("1. View expense summary")
        print("2. View goals progress")
        print("3. Add new expense")
        print("4. Budget check")
        print("5. Exit")

        choice = input("\nChoose option (1-5): ")

        if choice == "1":
            # Display expense summary
            print("\n--- Expense Summary ---")
            if len(expenses) == 0:
                print("No expenses recorded yet.")
            else:
                for i in range(len(expenses)):
                    print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
                print(f"\nTotal expenses: ${sum(expenses):.2f}")

        elif choice == "2":
            # Display goals progress
            print("\n--- Goals Progress ---")
            if len(goal_names) == 0:
                print("No savings goals set yet.")
            else:
                for i in range(len(goal_names)):
                    goal_name = goal_names[i]
                    target = goal_targets[i]
                    current = goal_current[i]
                    progress_percentage = (current / target) * 100 if target > 0 else 0
                    print(f"{goal_name}: ${current:.2f}/${target:.2f} ({progress_percentage:.1f}% complete)")

        elif choice == "3":
            # Add single new expense
            print("\n--- Add New Expense ---")

            while True:
                expense_input = input("Enter new expense amount: $")
                try:
                    expense_amount = float(expense_input)
                    if expense_amount <= 0:
                        print("Expense amount must be positive.")
                        continue
                    break
                except ValueError:
                    print("Invalid input. Please enter a valid number.")

            expense_description = input("What was this expense for? ")
            while not expense_description:
                print("Description cannot be empty. Please try again.")
                expense_description = input("What was this expense for? ")

            expenses.append(expense_amount)
            descriptions.append(expense_description)
            total_spent += expense_amount # Update global total_spent

            budget_percentage_used = (total_spent / monthly_budget) * 100
            print(f"Expense added! Current total spent: ${total_spent:.2f} ({budget_percentage_used:.1f}% of budget)")

        elif choice == "4":
            # Budget status check
            print("\n--- Budget Check ---")
            budget_percentage_used = (total_spent / monthly_budget) * 100
            remaining_budget = monthly_budget - total_spent

            print(f"Monthly Budget: ${monthly_budget:.2f}")
            print(f"Total Spent: ${total_spent:.2f}")
            print(f"Budget Used: {budget_percentage_used:.1f}%")
            print(f"Remaining Budget: ${remaining_budget:.2f}")

            if remaining_budget < 0:
                print("🚨 Warning! You are over budget! Consider cutting back on non-essential spending.")
            elif budget_percentage_used > 80:
                print("⚠️ You are close to your budget limit. Spend wisely for the rest of the month.")
            else:
                print("👍 Your spending is well within budget. Keep up the good work!")

        elif choice == "5":
            # Exit with summary
            running = False
            print(f"\nGoodbye {name}!")
            print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
            print("🚀 Your finance tracker is growing stronger each week!")

        else:
            print("Invalid choice! Please choose 1-5.")

    print("\n=== Finance Tracker Session Complete ===")

    # --- End of Combined Finance Tracker Code ---

finally:
    # Restore original stdin and stdout
    sys.stdin = original_stdin
    sys.stdout = original_stdout

# Print the captured output to the notebook
print(captured_output.getvalue())
```
